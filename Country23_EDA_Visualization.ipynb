#### 3. Poverty Visualization 3 - Pareto and pie chart of average poverty by region

```python
import matplotlib.pyplot as plt

# Aggregate: Mean %pov per region
df_pov_pie = df_pov_2.groupby('region')['%pov'].mean().sort_values(ascending=False)

# Create figure with two subplots
fig, axes = plt.subplots(2, 1, figsize=(8, 10))

# Pareto bar chart (left)
axes[0].bar(df_pov_pie.index, df_pov_pie.values, color='saddlebrown', label='Mean % Poverty')
axes[0].set_xlabel("Region")
axes[0].set_ylabel("Mean % Poverty")
axes[0].set_title("Pareto Chart of Mean % Poverty by Region")
axes[0].tick_params(axis='x', rotation=45)

# Pie chart (right)
axes[1].pie(df_pov_pie.values, labels=df_pov_pie.index, autopct='%1.1f%%', colors=plt.cm.Set2.colors)
axes[1].set_title("Pie Chart of Mean % Poverty by Region")

# Overall title
fig.suptitle("Pareto / Bar Chart of Mean % Poverty by Region (from regions represented in study)", fontsize=14)

plt.tight_layout()
plt.show()
```

#### 5. Poverty Visualization 5 - Scatter plot of poverty versus CPIA scores

#### 10. Poverty Visualization 10 - Correlation heatmap of transformed features in poverty dataset

```python
# Ensure df_pov_continous is assigned

# Compute the correlation matrix
corr_matrix_df2 = df_pov_2_transf.drop(columns='%pov')
corr_matrix_pov_transf = corr_matrix_df2.corr()

# Generate the heatmap
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix_pov_transf, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Transformed Features in Poverty Dataset")
plt.show()
```

The graph above shows the the following colinearity for both transformed and non-transformed poverty dataset
- cpia_inclusion is correlated with most of the other cpia features
- cpia_regulation and cpia_trade
- all of the income related features are correlated with each other
- imports and exports are correlated
Based on this the following features will be dropped
- cpia_inclusion
- cpia_trade
- income_quintie2
- income_quintile3
- income_quintile4
- income_quintile5
- income_top10%
- income_top-mid60
- comm_import_capita

Note: there does appear to be colinearity b/w college enrollment and exports on the transformed dataset, but both will stay in the study for now due to no obvious functional relationship

#### 22. Create Plotly Bubble plot for Poverty

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
import plotly.io as pio
pio.renderers.default = "notebook_connected"
```

```python
# Group your data by region to get the average poverty percentage
region_avg = df_pov_2.groupby("region")["%pov"].mean().reset_index()

# Add latitude and longitude from the dictionary
region_avg["lon"] = region_avg["region"].map(lambda r: region_centers.get(r, (None, None))[0])
region_avg["lat"] = region_avg["region"].map(lambda r: region_centers.get(r, (None, None))[1])
```

```python
# Bubble plot of poverty
fig = px.scatter_geo(
    region_avg,
    lat="lat",
    lon="lon",
    size="%pov",
    color="%pov",
    color_continuous_scale=["blue", "red"],  # NEW COLOR SCALE
    hover_name="region",
    projection="natural earth",
    title="Average Poverty by Region"
)

# White background, clean continents, blue ocean (optional)
fig.update_geos(
    showcountries=True,
    showland=True,
    landcolor="white",       # ALL LAND = WHITE
    showocean=True,
    oceancolor="wheat",      # OCEAN ALSO WHITE (or pick light blue if preferred)
)

fig.update_layout(
    geo=dict(
        showframe=False,
        showcoastlines=True,
    ),
    plot_bgcolor="white",     # Background white
    paper_bgcolor="white"
)

fig.show()

```
