#### 1.1 Collect poverty data

```python
df_gdp_3_not_transformed = pd.read_csv('https://raw.githubusercontent.com/te-ex153/Projects/refs/heads/main/df_gdp_3_not_transformed.csv')
df_gdp_3_transformed = pd.read_csv('https://raw.githubusercontent.com/te-ex153/Projects/refs/heads/main/df_gdp_3_transformed.csv')
```

```python
print('First five records of non-tranformed dataframe')
print(df_gdp_3_not_transformed.head(), '\n\n')

print('First five records of tranformed dataframe')
print(df_gdp_3_transformed.head())
```

#### 1.2 Data Preprocessing - Separate feature and target variables
[üîù Return to Table of Contents](#Table-of-Contents)

```python
X_nt = df_gdp_3_not_transformed.drop(columns='gdp')
y_nt = df_gdp_3_not_transformed[['gdp']]
print('First five records of non-transformed target data')
print(y_nt.head(), '\n')
print('First five records of non-transformed feature data')
print(X_nt.head(), '\n')
X_t = df_gdp_3_transformed.drop(columns='gdp')
y_t = df_gdp_3_transformed[['gdp']]
print('First five records of transformed target data')
print(y_t.head(),'\n')
print('First five records of transformed feature data')
print(X_t.head())
```

#### 1.3 Data Preprocessing - Normalize feature variables for transformed and non-transformed data
[üîù Return to Table of Contents](#Table-of-Contents)

```python
# Set scaler
scaler = StandardScaler()

# Fit and transform X_t and X_nt
X_t_scaled = scaler.fit_transform(X_t)
X_nt_scaled = scaler.transform(X_nt)  # Use transform to keep the same scaling

# Convert back to DataFrame if needed
import pandas as pd
X_t_scaled = pd.DataFrame(X_t_scaled, columns=X_t.columns, index=X_t.index)
X_nt_scaled = pd.DataFrame(X_nt_scaled, columns=X_nt.columns, index=X_nt.index)
```

#### 1.4 Data Preprocessing - Convert transformed continuous target variable into binary and multi-class variables
[üîù Return to Table of Contents](#Table-of-Contents)

```python
# Ensure y_t is a 1D Pandas Series (flatten if necessary)
y_t = y_t.squeeze()  

# Compute log values for thresholding
log_7k, log_22k = np.log(7000), np.log(22000)  
log_12k = np.log(12000)  

# Convert y_t (log-transformed) into three classes using log(15) and log(30)
bins_3_t = [float('-inf'), log_7k, log_22k, float('inf')]  
labels_3_t = [0, 1, 2]  
y_t_cat3 = pd.cut(y_t, bins=bins_3_t, labels=labels_3_t)

# Convert y_t (log-transformed) into two classes using log(20)
bins_2_t = [float('-inf'), log_12k, float('inf')]  
labels_2_t = [0, 1]  
y_t_cat2 = pd.cut(y_t, bins=bins_2_t, labels=labels_2_t)

# Display class distributions
print("y_t_cat3 distribution:", y_t_cat3.value_counts())
print("y_t_cat2 distribution:", y_t_cat2.value_counts())
```

#### 1.5 Data Preprocessing - Convert continuous non-transformed target variable into binary and multi-class variables
[üîù Return to Table of Contents](#Table-of-Contents)

```python
# Ensure y_nt is a 1D Pandas Series (flatten if necessary)
y_nt = y_nt.squeeze()  

# Convert y_nt into three classes using thresholds 15 and 30
bins_3_nt = [float('-inf'), 7000, 22000, float('inf')]  
labels_3_nt = [0, 1, 2]  
y_nt_cat3 = pd.cut(y_nt, bins=bins_3_nt, labels=labels_3_nt)

# Convert y_nt into two classes using threshold 20
bins_2_nt = [float('-inf'), 12000, float('inf')]  
labels_2_nt = [0, 1]  
y_nt_cat2 = pd.cut(y_nt, bins=bins_2_nt, labels=labels_2_nt)

# Display class distributions
print("y_nt_cat3 distribution:", y_nt_cat3.value_counts(), '\n')
print("y_nt_cat2 distribution:", y_nt_cat2.value_counts())
```

#### 1.6 Develop functions for heatmap confusion matrix to be used in model evaluation
[üîù Return to Table of Contents](#Table-of-Contents)

```python
# Function for Multi-class target variable
def plot_confusion_matrix(y,y_predict):
    "this function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix'); 
    ax.xaxis.set_ticklabels(['Low GDP', 'Moderate GDP', 'High GDP']); ax.yaxis.set_ticklabels(['Low GDP', 'Moderate GDP', 'High GDP']) 
    plt.show() 
    
    # Function for 2-class target variable
def plot_confusion_matrix2(y,y_predict):
    "this function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix'); 
    ax.xaxis.set_ticklabels(['Lower GDP', 'Higher GDP']); ax.yaxis.set_ticklabels(['Lower GDP', 'Higher GDP']) 
    plt.show() 
```

### 2. Regression Models

#### 2.1 Regression Model's train/test split of continuous transformed dataset
[üîù Return to Table of Contents](#Table-of-Contents)

```python
# Split training/testing dataset 
X_train, X_test, y_train, y_test = train_test_split(X_t_scaled, y_t, test_size=0.25, random_state=0)
```

#### 2.2 Linear & Polynomial Regression - Train and fit data
[üîù Return to Table of Contents](#Table-of-Contents)

```python
# Define the pipeline with PolynomialFeatures and LinearRegression
pipeline = Pipeline([
    ('poly', PolynomialFeatures()),  # Placeholder for different degrees
    ('linear', LinearRegression())
])

# Define hyperparameter grid
param_grid = {
    'poly__degree': [1, 2, 3, 4, 5],  # Try polynomial degrees from 1 to 5
    'linear__fit_intercept': [True, False]
}

# Set up GridSearchCV
grid_search_poly = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=5,  # 5-fold cross-validation
    scoring='r2',  # Optimize for R¬≤ score
    n_jobs=-1
)

# Fit to the training data
grid_search_poly.fit(X_train, y_train)

# Print best parameters and best score
print("Best Polynomial Degree:", grid_search_poly.best_params_['poly__degree'])
print("Best R¬≤ Score:", grid_search_poly.best_score_)
```

##### 2.2a Linear & Polynomial Regression - Evaluate model performance metrics

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Get the best model from GridSearchCV
best_poly_reg = grid_search_poly.best_estimator_

# Predict on the test set
y_pred = best_poly_reg.predict(X_test)

# Calculate regression metrics
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)  # RMSE is sqrt of MSE
mape = (abs((y_test - y_pred) / y_test)).mean() * 100  # Expressed as %

# Print results
print(f"R¬≤ Score: {r2:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")

# Plot residuals
residuals = y_test - y_pred
plt.figure(figsize=(8, 6))
sns.histplot(residuals, bins=30, kde=True)
plt.axvline(0, color='red', linestyle='dashed')
plt.title("Residual Distribution")
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.show()
```

#### 2.2a Linear & Polynomial Regression - Evaluate feature importance ranking

```python
# Get the best Polynomial Regression model from GridSearchCV
best_poly_reg = grid_search_poly.best_estimator_

# Extract feature names (if X_train is a DataFrame)
poly = best_poly_reg.named_steps['poly']  # Access PolynomialFeatures
feature_names = poly.get_feature_names_out(input_features=X_train.columns)  # Get expanded feature names

# Extract model coefficients
coef = best_poly_reg.named_steps['linear'].coef_

# Sort features by absolute importance
sorted_idx = np.argsort(abs(coef))  # Sort by magnitude

# Plot feature importance
plt.figure(figsize=(8, 6))
plt.barh(np.array(feature_names)[sorted_idx], coef[sorted_idx], color="blue")
plt.xlabel("Coefficient Value (Feature Importance)")
plt.ylabel("Features")
plt.title("Feature Importance for Polynomial Regression")
plt.show()
```

#### 2.3 Support Vector Machine Regressor - Train and fit data
[üîù Return to Table of Contents](#Table-of-Contents)

```python
# Define the pipeline with StandardScaler and SVR
pipeline_svr = Pipeline([
    ('scaler', StandardScaler()),  # StandardScaler for feature scaling
    ('svr', SVR())  # Support Vector Regression
])

# Define hyperparameter grid
param_grid_svr = {
    'svr__kernel': ['linear', 'rbf', 'poly'],  # Different kernels
    'svr__C': [0.1, 1, 10, 100],  # Regularization parameter
    'svr__epsilon': [0.01, 0.1, 0.2],  # Epsilon parameter for the SVR model
}

# Set up GridSearchCV
grid_search_svr = GridSearchCV(
    estimator=pipeline_svr,
    param_grid=param_grid_svr,
    cv=5,  # 5-fold cross-validation
    scoring='r2',  # Optimize for R¬≤ score
    n_jobs=-1
)

# Fit to the training data
grid_search_svr.fit(X_train, y_train)

# Print best parameters and best score
print("Best Kernel:", grid_search_svr.best_params_['svr__kernel'])
print("Best C:", grid_search_svr.best_params_['svr__C'])
print("Best Epsilon:", grid_search_svr.best_params_['svr__epsilon'])
print("Best R¬≤ Score:", grid_search_svr.best_score_)

# Predict on the test set
y_pred_svr = grid_search_svr.predict(X_test)

# R¬≤ score (explained variance)
r2 = r2_score(y_test, y_pred_svr)

# Mean Squared Error
mse = mean_squared_error(y_test, y_pred_svr)

# Mean Absolute Error
mae = mean_absolute_error(y_test, y_pred_svr)

print(f"R¬≤: {r2:.4f}")
print(f"MSE: {mse:.4f}")
print(f"MAE: {mae:.4f}")

# Calculate residuals
residuals = y_test - y_pred_svr

# Create a distribution plot for residuals
plt.figure(figsize=(8, 6))
sns.histplot(residuals, kde=True, color='blue', bins=30, stat='density')
plt.axvline(x=0, color='black', linestyle='--')  # Zero line for reference
plt.xlabel('Residuals')
plt.ylabel('Density')
plt.title('Distribution of Residuals for Support Vector Regression')
plt.show()
```

##### 2.3a Support Vector Machine Regressor - Evaluate feature importance ranking

```python
# Get the best SVR model from GridSearchCV
best_svr_model = grid_search_svr.best_estimator_

# Compute permutation importance
result = permutation_importance(
    best_svr_model, X_test, y_test, scoring='r2', n_repeats=10, random_state=42
)

# Extract importance values
feature_importance = result.importances_mean  

# Get feature names
feature_names = X_train.columns if hasattr(X_train, 'columns') else [f"Feature {i}" for i in range(X_train.shape[1])]

# Sort and plot
sorted_idx = np.argsort(feature_importance)
plt.figure(figsize=(8, 6))
plt.barh(np.array(feature_names)[sorted_idx], feature_importance[sorted_idx], color="blue")
plt.xlabel("Permutation Importance")
plt.ylabel("Features")
plt.title("Feature Importance (Permutation) for SVR")
plt.show()
```
